```python
#處理speed01

data01=pd.read_csv("traff_flow_speed01.csv",index_col=0)
data01["車速*通行量"]=data01["平均車速"]*data01["通行量"]
```

```python
da01=data01.groupby(by=["日期", "偵測站編號-起","偵測站編號-迄"])["通行量"].sum()
da01=pd.DataFrame(da01)
da01.rename(columns={"通行量":"總通行量"},inplace=True)
da01.reset_index(inplace=True)
```

```python
df01=data01.groupby(by=["日期", "偵測站編號-起","偵測站編號-迄"])["車速*通行量"].sum()/data01.groupby(by=["日期", "偵測站編號-起","偵測站編號-迄"])["通行量"].sum()
df01=pd.DataFrame(df01)
df01.rename(columns={0:"總平均車速"},inplace=True)
df01.reset_index(inplace=True)
```

```python
mer01=pd.merge(da01,df01,on=["日期","偵測站編號-起","偵測站編號-迄"])
mer01["偵測站編號-起"]=[i.replace("R","0") for i in mer01["偵測站編號-起"]]
mer01["偵測站編號-迄"]=[i.replace("R","0") for i in mer01["偵測站編號-迄"]]
mer01["起"]=[int(i[3:7])/10 for i in mer01["偵測站編號-起"]]
mer01["迄"]=[int(i[3:7])/10 for i in mer01["偵測站編號-迄"]]
mer01["日期"]=[datetime.strptime(t,"%Y/%m/%d %H:%M") for t in mer01["日期"]] 
mer01["國道向"]=[(i[1:2]+i[7:8]) for i in  mer01["偵測站編號-起"]]
mer01
```
