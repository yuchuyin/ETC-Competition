```python
data=pd.read_csv("C:/Users/user/Desktop/ETC/accident/accident108.csv")
data.head(5)
```

```python
print(data.shape)
print(data.columns)
print(data.isnull().sum())
```

```python
data["日期"]=[datetime.strptime(t,"%Y/%m/%d") for t in data["日期"]] 
data["月"]=[m.month for m in data["日期"]]
data["日"]=[m.day for m in data["日期"]]
data["星期"]=[calendar.weekday(2019, i.month, i.day) for i in data["日期"]]
```

```python
a_list=[]
for i in data["道路型態(子)"]:
    if i<3:
        a=1
    elif 2<i<6:
        a=2
    elif 5<i<16:
        a=3
    else:
        a=4
    a_list.append(a)
data["道路型態(主)"]=a_list
        
```

```python
b_list=[]
for i in data["事故位置(子)"]:
    if i <5:
        b=1
    elif 4<i<14:
        b=2
    elif 13<i<18:
        b=3
    else:
        b=4
    b_list.append(b)
data["事故位置(主)"]=b_list
```

```python
c_list=[]
for i in data["車道劃分設施-分向設施(子)"]:
    if i <4:
        c=1
    elif 3<i<6:
        c=2
    elif 5<i<8:
        c=3
    elif 7<i<10:
        c=4
    else:
        c=5
    c_list.append(c)
data["車道劃分設施-分向設施(主)"]=c_list
```

```python
d_list=[]
for i in data["事故類型及型態(子)"]:
    if i <10:
        d=1
    elif 9<i<18:
        d=2
    elif 17<i<30:
        d=3
    else:
        d=4
    d_list.append(d)
data["事故類型及型態(主)"]=d_list
```

```python
data["肇因研判(子)"]=[i.replace('  ',"44") for i in data["肇因研判(子)"]]    
data["肇因研判(子)"]=[int(i) for i in data["肇因研判(子)"]]    
```

```python
e_list=[]
for i in data["肇因研判(子)"]:
    if i <27:
        e=1
    elif 26<i<30:
        e=2
    elif 29<i<38:
        e=3
    elif 37<i<44:
        e=4
    elif i==44:
        e=5
    elif 44<i<51:
        e=6
    elif 50<i<61:
        e=7
    elif 60<i<66:
        e=8
    else:
        e=9
    e_list.append(e)
data["肇因研判(主)"]=e_list
```

```python
data["當事者車種(子)"]=[i.replace('  ',"None") for i in data["當事者車種(子)"]]    
f_list=[]
for i in data["當事者車種(子)"]:
    if i in ["A01","A02","A03","A04","A05","A06"]:
        f=1
    elif i in ["A11","A12"]:
        f=2
    elif i in ["A21","A22"]:
        f=3
    elif i in ["A31","A32"]:
        f=4
    elif i in ["A41","A42"]:
        f=5
    elif i in ["B01","B02","B03"]:
        f=6
    elif i in ["B11","B12"]:
        f=7
    elif i in ["C01","C02","C03","C04","C05"]:
        f=8
    elif i in ["D01","D02","D03"]:
        f=9
    elif i in ["E01","E02","E03","E04","E05"]:
        f=10
    elif i in ["F01","F02","F03","F04","F05","F06"]:
        f=11
    elif i in ["G01","G02","G03","G04","G05","G06"]:
        f=12
    elif i in ["H01","H02","H03"]:
        f=13
    else:
        f=0
    f_list.append(f)
data["當事者車種(主)"]=f_list
```

```python
data=data[['日期','月','日', '時間', "星期",'交通事故類別', '縣市', '市區鄉鎮', '公路名稱', '路線公里', '路線公尺', '向車道',
       '24小時內死亡人數', '2-30日內死亡人數', '受傷人數', '天候代碼', '光線代碼', '道路類別', '速限','道路型態(主)',
       '道路型態(子)','事故位置(主)' ,'事故位置(子)', '路面狀況-鋪裝', '路面狀況-狀態', '路面狀況-缺陷', '道路障礙-障礙物',
       '道路障礙-視距', '號誌-號誌種類', '號誌-號誌動作', '車道劃分設施-分向設施(主)','車道劃分設施-分向設施(子)',
       '車道劃分設施-分道設施-快車道或一般車道間', '車道劃分設施-分道設施-快慢車道間', '車道劃分設施-分道設施-路面邊線',
       '事故類型及型態(主)','事故類型及型態(子)','當事者車種(主)','當事者車種(子)','肇因研判(主)','肇因研判(子)']]
```
